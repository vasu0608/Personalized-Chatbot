```python
import pandas as pd
from datasets import Dataset, DatasetDict
from sklearn.model_selection import train_test_split
df = pd.read_excel("/content/Personalized LLM database.xlsx")
df = df[["Question", "Personalized Response"]].dropna()
df = df.rename(columns={
    "Question": "instruction",
    "Personalized Response": "response"
})

train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

dataset = DatasetDict({
    "train": Dataset.from_pandas(train_df),
    "test": Dataset.from_pandas(test_df)
})

print(dataset)
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

model_name = "distilgpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)

tokenizer.pad_token = tokenizer.eos_token

model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
def format_function(example):
    text = ["### Question:\n" + instruction + "\n\n### Answer:\n" + response for instruction, response in zip(example["instruction"], example["response"])]
    return tokenizer(
        text,
        truncation=True,
        padding="max_length",
        max_length=256
    )

tokenized_dataset = dataset.map(format_function, batched=True)
```

```python
import os
os.environ["WANDB_DISABLED"] = "true"

from transformers import TrainingArguments, Trainer, DataCollatorForLanguageModeling

training_args = TrainingArguments(
    output_dir="./results",
    eval_strategy="epoch",
    learning_rate=5e-5,
    per_device_train_batch_size=2,
    per_device_eval_batch_size=2,
    num_train_epochs=2,
    weight_decay=0.01,
    save_total_limit=2,
    logging_dir="./logs",
    logging_steps=10,
    push_to_hub=False,
    report_to="none"
)

data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset["train"],
    eval_dataset=tokenized_dataset["test"],
    tokenizer=tokenizer,
    data_collator=data_collator
)
```

```python
import torch

new_questions = [
    "What is the best way to learn a new programming language?",
    "How does a self-driving car work?",
    "What are the benefits of meditation?"
]

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

for question in new_questions:
    input_text = "### Question:\n" + question + "\n\n### Answer:\n"
    input_ids = tokenizer.encode(input_text, return_tensors="pt").to(device)

    output = model.generate(
        input_ids,
        max_length=200,
        num_return_sequences=1,
        no_repeat_ngram_size=2,
        do_sample=True,
        top_k=50,
        top_p=0.95,
        temperature=0.7
    )

    generated_response = tokenizer.decode(output[0], skip_special_tokens=True)

    response_start = generated_response.find("### Answer:\n")
    if response_start != -1:
        generated_answer = generated_response[response_start + len("### Answer:\n"):].strip()
    else:
        generated_answer = "Could not generate a response."

    print(f"Question: {question}")
    print(f"Generated Response: {generated_answer}\n")
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

model_path = "./results"
tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForCausalLM.from_pretrained(model_path)

chatbot = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer
)

print("🤖 Chatbot ready! Type 'quit' to exit.\n")

while True:
    user_input = input("You: ")
    if user_input.lower() in ["quit", "exit", "bye"]:
        print("Bot: Goodbye 👋")
        break

    prompt = f"Question: {user_input}\nAnswer:"

    response = chatbot(
        prompt,
        max_new_tokens=300,
        do_sample=True,
        top_p=0.95,
        temperature=0.8,
        repetition_penalty=1.5,
        eos_token_id=tokenizer.eos_token_id
    )


    generated_text = response[0]["generated_text"]
    answer_start = generated_text.find("Answer:")
    if answer_start != -1:

        answer = generated_text[answer_start + len("Answer:"):].strip()

        next_question_start = answer.find("Question:")
        if next_question_start != -1:
            answer = answer[:next_question_start].strip()
    else:
        answer = "Could not generate a relevant response."

    print("Bot:", answer)
```
